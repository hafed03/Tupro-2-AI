```
#rumus jarak L2(euclidean)
def euclidean(trainning,testing):
  jarak=[]
  for i in range(len(trainning)-1):
    if trainning[i] != testing[i]:
      jarak.append(math.sqrt((trainning[i]-testing[i])**2))
  return sum(jarak)
```

```
def knn(k,trainning,testing):
  knn = []
  for i in range(len(trainning)):
    knn.append([euclidean(trainning[i],testing), trainning[i][-1]])
  knn.sort()
  return knn[:k]
```

```
def prediksi(k,trainning,testing):
  prediksi = []
  for i in range(len(testing)):
    temp_knn = knn(k,trainning,testing[i])
    unik = Counter([temp_knn[i][1] for i in range(len(temp_knn))])
    prediksi.append([temp_knn[i] for i in range(len(temp_knn)) if temp_knn[i][1] == max(unik, key=unik.get)][0])
  return prediksi
```

```
def akurasi(testing,prediksi):
  akur = 0
  for i in range(len(prediksi)):
    if prediksi[i][-1]==testing[i][-1]:
      akur +=1
    else:
      akur +=0  
  return ((akur/len(testing)) *100)
```

```
list_k = []
K = 20
for i in range(K):
  list_ac = []
  k = i+1
  for i in range(len(train)):
    list_ac.append(akurasi(test[i],prediksi(k,train[i],test[i])))
  avg = sum(list_ac)/5
  list_k.append(avg)
max(list_k)
for i in range(len(list_k)):
  if max(list_k) == list_k[i]:
    temp = i+1    
print("Nilai k terbaik :",temp)   
print("Rata-rata akurasi :",max(list_k)) 
```
